### Demo 1: Matrix Inversion is Unstable

```python
import numpy as np
from scipy.linalg import hilbert
from sklearn.linear_model import LinearRegression

np.set_printoptions(linewidth=10000, suppress=True)
```

```python
np.random.seed(2109)
A = np.random.rand(12, 12)
print(A)
```

```python
A_inv = np.linalg.inv(A)
print(A @ A_inv)
```

```python
A = hilbert(12)
print(A)
```

```python
A_inv = np.linalg.inv(A)
print(A @ A_inv)
```

### Demo 2: Ill-conditioned Matrices in Linear Regression

```python
# Create a dataset with almost linearly dependent features.
# x2 = x1 + epsilon * noise

epsilon = 1e-12   # try 1e-6, 1e-10, 1e-12 to make it worse

np.random.seed(2109)

x1 = np.random.randn(100, 1)
x2 = x1 + epsilon * np.random.randn(100, 1)

X = np.hstack([x1, x2])

# True underlying weights (well-behaved)
w_true = np.array([1.0, -1.0])

y = X @ w_true + 0.01 * np.random.randn(100)
```

```python
lr = LinearRegression(fit_intercept=False)
lr.fit(X, y)
w_sklearn = lr.coef_

print(f"True weights:                               {w_true}")
print(f"scikit-learn LinearRegression weights:      {w_sklearn}")
```
